```python
import pandas as pd

# 1. SAS 포맷인 XPT 파일을 읽어옵니다. (파일명이 정확해야 합니다)
df = pd.read_sas('DEMO_J.XPT')

# 2. 데이터가 잘 들어왔는지 상위 5개만 눈으로 확인합니다.
print("축하합니다! 미국 NHANES 데이터 로딩 성공!")
df.head()
```

```python
# 1. 통증 설문 데이터(PAIN_J.XPT)를 불러옵니다.
df_pain = pd.read_sas('WHQ_J.XPT')

# 2. 인구학 데이터(df)와 통증 데이터(df_pain)를 'SEQN'(고유 환자 번호) 기준으로 합칩니다.
# how='inner'는 두 데이터에 모두 존재하는 사람만 남기겠다는 뜻입니다 (교집합).
df_merged = pd.merge(df, df_pain, on='SEQN', how='inner')

# 3. 데이터가 잘 합쳐졌는지 확인합니다.
print(f"축하합니다! 총 {len(df_merged)}명의 통합 데이터셋이 준비되었습니다.")
df_merged.head()
```

```python
import matplotlib.pyplot as plt

# 1. 분석할 변수 선택: WHQ030 (체중 인식)
# 데이터 중에 응답 거부(7)나 모름(9)을 제외하고, 1,2,3번 응답자만 추립니다.
df_clean = df_merged[df_merged['WHQ030'].isin([1, 2, 3])].copy()

# 2. 숫자를 사람이 읽기 쉬운 라벨로 변경 (Mapping)
# 이것이 데이터 전처리의 핵심입니다.
label_map = {
    1: 'Overweight (과체중)',
    2: 'Underweight (저체중)',
    3: 'About Right (적정)'
}
df_clean['Perception_Label'] = df_clean['WHQ030'].map(label_map)

# 3. 각 응답별 인원수 세기
counts = df_clean['Perception_Label'].value_counts()

# 4. 파이 차트 그리기
plt.figure(figsize=(8, 8))  # 그래프 크기 설정
plt.pie(counts, 
        labels=counts.index, 
        autopct='%1.1f%%',  # 퍼센트 표시 (소수점 첫째 자리)
        startangle=90,      # 시작 각도
        colors=['#ff9999','#66b3ff','#99ff99']) # 색상 예쁘게
plt.title('How do US adults perceive their weight? (NHANES 2017-2018)', fontsize=15)
plt.show()
```

```python
import statsmodels.formula.api as smf
import numpy as np

# 1. 분석을 위한 데이터 준비
# WHQ030(체중 인식)이 1번(Overweight)이면 1, 아니면 0인 'Target' 변수를 만듭니다.
# (이것이 우리가 예측하려는 종속변수 Y입니다)
df_merged['is_overweight'] = np.where(df_merged['WHQ030'] == 1, 1, 0)

# 성별(RIAGENDR) 변수 처리: 1=남성, 2=여성 -> 분석하기 쉽게 0=남성, 1=여성으로 변경
df_merged['is_female'] = np.where(df_merged['RIAGENDR'] == 2, 1, 0)

# 2. 로지스틱 회귀분석 모델 설계 (R 스타일의 포뮬러 방식)
# "과체중 인식 여부(Y)는 나이(Age)와 성별(Female)에 의해 결정된다"
model = smf.logit('is_overweight ~ RIDAGEYR + is_female', data=df_merged)

# 3. 모델 학습 및 결과 요약
result = model.fit()
print(result.summary())

# 4. 오즈비(Odds Ratio) 계산 (로그 스케일을 보기 쉽게 변환)
print("\n[Odds Ratios (오즈비)]")
print(np.exp(result.params))
```

```python
# 1. 신체 계측 데이터(BMX_J.XPT) 불러오기
df_body = pd.read_sas('BMX_J.XPT')

# 2. 기존 데이터(df_merged)에 합치기
# 이미 합쳐진 데이터에 또 합치는 것을 'Chain Merge'라고 합니다.
df_final = pd.merge(df_merged, df_body, on='SEQN', how='inner')

# 3. BMI 변수 확인 (NHANES 변수명: BMXBMI)
# 결측치(NaN)가 있으면 에러가 날 수 있으니 제거합니다.
df_final = df_final.dropna(subset=['BMXBMI', 'WHQ030', 'RIDAGEYR', 'RIAGENDR'])

# 4. 새로운 모델 설계 (Model 2)
# "과체중 인식(Y) ~ 나이 + 성별 + *실제 BMI*"
model_adj = smf.logit('is_overweight ~ RIDAGEYR + is_female + BMXBMI', data=df_final)
result_adj = model_adj.fit()

# 5. 결과 비교 (Model 1 vs Model 2)
print(result_adj.summary())

print("\n[Model 2: Adjusted Odds Ratios (BMI 보정 후)]")
print(np.exp(result_adj.params))
```
